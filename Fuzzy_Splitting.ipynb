```python
import numpy as np
import pandas as pd
import statistics
import pycaret
from pycaret.classification import *
import matplotlib.pyplot as plt
```

## WESAD

```python
wesad_dataset = pd.read_csv('Final_CSVs/wesad_fuzzy4_12.csv')
```

```python
wesad_dataset
```

```python
# Drop 'dataset' to run pycaret tests based on "Cluster".

wesad_dataset = wesad_dataset.drop('dataset', axis = 1)
wesad_dataset
```

```python
unique_participants = wesad_dataset["Cluster"].unique()
wesad_group = wesad_dataset.groupby('Cluster')
```

```python
accuracies = []
precision = []
recall = []
f1scores = []

for participant in unique_participants:
    print("Participant: ",participant)
    part_df = wesad_group.get_group(participant)
    grid = setup(data=part_df, target='stress', html=False, silent=True, verbose=False) #fix_imbalance = True,
    best = compare_models(sort="F1")
    accuracies.append(pull()['Accuracy'][0])
    precision.append(pull()['Prec.'][0])
    recall.append(pull()['Recall'][0])
    f1scores.append(pull()['F1'][0])
    print(best)
```

```python
mean_acc = statistics.mean(accuracies)
mean_prec = statistics.mean(precision)
mean_rec = statistics.mean(recall)
mean_f1 = statistics.mean(f1scores)
```

```python
print("Mean Accuracy WESAD - Fuzzy Clustering: ", mean_acc)
print("Mean Precision WESAD - Fuzzy Clustering: ", mean_prec)
print("Mean Recall WESAD - Fuzzy Clustering: ", mean_rec)
print("Mean F1-score WESAD - Fuzzy Clustering: ", mean_f1)
```
